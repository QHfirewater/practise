## 绘图的基本用法

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#生成数据
x = np.linspace(-3,3,50)
y1 = 2*x +1
y2 = x**2
```

```python
# plt.style.use('ggplot')
plt.figure(num = 3,figsize=(8,5))#设置画纸，num 是画纸的编号
plt.plot(x,y1,'b')
plt.plot(x,y2,linestyle='--',color = 'r',linewidth = '1')
plt.xlim([-1,2])    #设置坐标最大值和最小值
plt.ylim([-4,4])
plt.xlabel('l am x')  #设置坐标便签
plt.ylabel('I am y')


new_tick = np.linspace(-1,2,5)
plt.xticks(new_tick)    #设置坐标分布
plt.yticks([-2,-1.8,-1,1.22,3],['really bad','bad','normal','good','really goog'])  #设置坐标伪标签分布


ax = plt.gca() #获取当前活动的axes
ax.spines['right'].set_color('none')    #取消右边测边框
ax.spines['top'].set_color('red')       #设置上边边框颜色
ax.xaxis.set_ticks_position('bottom')   #设置刻度线所在位置

ax.spines['bottom'].set_position(('data',0))#通过设置边框，间接设置刻度线位置
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
```

```python
x = np.linspace(-3,10,50)
y = x-1
```

```python
plt.figure()
plt.plot(x,y,linestyle='--',linewidth=1)
plt.xlim([-1,2])
plt.ylim([-2,1])
plt.xticks([-1,-0.5,1],['bad','normal','good'])
ax = plt.gca()
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
```

## 设置图例和标注

```python
x = np.linspace(-3,3,50)
y1 = 2*x +1
y2 = x**2
```

```python
fig = plt.figure()
ax = fig.add_subplot()
line1 = ax.plot(x,y1,color = 'b',label = 'linear line')
line2 = ax.plot(x,y2,color = 'r',linewidth = 1,linestyle = '--',label = 'square line')
# ax.legend(loc = 'upper center') #直接设置图例的方法
ax.legend(labels=['1 line','2 line'],loc = 'upper center') #灵活设置图例的方法
```

```python
x = np.linspace(-3, 3, 50)
y = 2*x + 1
plt.figure(num=1, figsize=(8, 5),)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
plt.plot(x, y,)

#通过绘制图形，显示图解
x0 = 1
y0 = 2*x0 + 1
plt.plot([x0, x0,], [0, y0,], 'k--', linewidth=2.5)
# set dot styles
plt.scatter([x0], [y0], s=50, color='b')


#设置箭头
#（xy表示箭头所在的位置，xycoords='data'说明是基于数据的值来选位置的，
# xytext=(+30, -30) 和 textcoords='offset points' 表示对于标注位置的描述 和 xy 偏差值，即标注位置是 xy 位置向右移动 30，向下移动30, 
# arrowprops是对图中箭头类型和箭头弧度的设置，需要用 dict 形式传入）
plt.annotate(r'2x+1=%s' % y0, xy=(x0, y0), xycoords='data', xytext=(+30, -30),
             textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))

plt.text(-3.7, 3, r'$This\ is\ the\ some\ text. \mu\ \sigma_i\ \alpha_t$',
         fontdict={'size': 16, 'color': 'r'})#-3.7,3, 是选取text的位置,
```

```python
plt.plot(x, y, linewidth=10, zorder=1)
plt.ylim(-2, 2)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

for label in ax.get_xticklabels() + ax.get_yticklabels(): #这里新建一个坐标轴，通过覆盖的方式将坐标轴显示出来
    label.set_fontsize(12)
    # 在 plt 2.0.2 或更高的版本中, 设置 zorder 给 plot 在 z 轴方向排序
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.7, zorder=2))
plt.show()
```

## 绘制其他类型的图形

### 散点图

```python
n = 1024    # data size
x = np.random.normal(0, 1, n) # 每一个点的X值
y = np.random.normal(0, 1, n) # 每一个点的Y值
t = np.arctan2(y,x) # for color value
```

```python
plt.scatter(x,y,s = 75,c= t,alpha =0.5)
plt.xlim(-1.5,1.5)
plt.xticks(())#设置忽略xtick
plt.ylim(-1.5,1.5)
plt.yticks(())
```

### 柱状图

```python
x = np.arange(12)
y = (1-x/float(12))*np.random.uniform(0.5,1,12)
```

```python
plt.bar(x,y)
plt.bar(x,-y)
plt.xlim(-0.5,12)
plt.ylim(-1.25,1,25)
```

```python
plt.figure()
plt.bar(x,y,facecolor = '#FFCCCC',edgecolor = 'white')
plt.bar(x,-y,facecolor = '#6699CC',edgecolor='white')
for m,n in zip(x,y):
    plt.text(m,n,'%.2f'%n,ha = 'center',va = 'bottom')
for m,n in zip(x,-y):
    plt.text(m,n,'%.2f'%n,ha = 'center',va = 'top')
```

### 等高线图

```python
def f(x,y):
    # the height function
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 -y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X,Y = np.meshgrid(x, y)
# use plt.contourf to filling contours
# X, Y and value for (X,Y) point
plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap=plt.cm.RdBu)
```

### 随机矩阵画图

```python
a = np.array([0.313660827978, 0.365348418405, 0.423733120134,
              0.365348418405, 0.439599930621, 0.525083754405,
              0.423733120134, 0.525083754405, 0.651536351379]).reshape(3,3)


#origin='lower'代表的就是选择的原点的位置。
#而 interpolation 表示画图方式，从该链接可看到matplotlib官网上对于内插法的不同方法的描述。
#这里我们使用的是内插法中的 Nearest-neighbor 的方法，其他的方式也都可以随意取选。
plt.imshow(a, interpolation='nearest', cmap='RdBu', origin='lower')
plt.colorbar(shrink=.92)
```

### subplot2grid规划图形

```python
plt.figure()
ax1 = plt.subplot2grid((3,3),(0,0),colspan =3)

#(3,3)表示将整个图像窗口分成3行3列, (1,0)表示从第1行第0列开始作图，colspan=2表示列的跨度为2
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
```

### gridspec规划图形
gridspec 同样能帮助我们画出前文的图，个人觉得 gridspec 的使用是最为方便的

```python
import matplotlib.gridspec as gridspec
plt.figure(figsize=(8, 8))
gs = gridspec.GridSpec(3, 3)


# gs[0, :]表示这个图占第0行和所有列, gs[1, :2]表示这个图占第1行和第2列前的所有列
ax6 = plt.subplot(gs[0, :])
ax7 = plt.subplot(gs[1, :2])
ax8 = plt.subplot(gs[1:, 2])
ax9 = plt.subplot(gs[-1, 0])
ax10 = plt.subplot(gs[-1, -2])
```

## 3D作图

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#生成坐标轴
fig = plt.figure()
ax = Axes3D(fig)

#生成数据集
# X, Y value
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)    # x-y 平面的网格
R = np.sqrt(X ** 2 + Y ** 2)
# height value
Z = np.sin(R)

#开始绘图
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
ax.contourf(X, Y, Z, zdir='z', offset=-1, cmap=plt.get_cmap('rainbow'))#数据集在xy平面上投影
```

如果 zdir 选择了x，那么效果将会是对于 XZ 平面的投影

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
ax.contourf(X, Y, Z, zdir='x', offset=-4, cmap=plt.get_cmap('rainbow'))
```

```python
from matplotlib import pyplot as plt
from matplotlib import animation
import numpy as np

fig, ax = plt.subplots()

x = np.arange(0, 2*np.pi, 0.01)
line, = ax.plot(x, np.sin(x))

def animate(i):
    line.set_ydata(np.sin(x + i/10.0))
    return line,
def init():
    line.set_ydata(np.sin(x))
    return line,

fig, ax = plt.subplots()
line, = ax.plot(x, np.sin(x))

ani = animation.FuncAnimation(fig=fig,
                              func=animate,
                              frames=100,
                              init_func=init,
                              interval=20,
                              blit=True)
plt.show()
```
